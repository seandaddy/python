```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('penguins')
```

```python
plt.scatter(df['flipper_length_mm'], df['body_mass_g'])
plt.xlabel('Flipper Length')
plt.ylabel('Body Mass')
plt.show()
```

```python
df_group = df.groupby('species')['body_mass_g'].mean().reset_index()
plt.bar(x=df_group['species'], height=df_group['body_mass_g'])
plt.xlabel('Species')
plt.ylabel('Body Mass')
plt.show()
```

```python
plt.hist(df['body_mass_g'], bins=30)
plt.xlabel('Body Mass')
plt.ylabel('Count')
plt.title('Weight Distribution of Penguins')
plt.show()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 6))

# Figure 1
axes[0].scatter(df['flipper_length_mm'], df['body_mass_g'])
axes[0].set_xlabel('Flipper Length(mm)')
axes[0].set_ylabel('Body Mass(g)')
axes[0].set_title('Correlation between Flipper and Body Mass')

# Figure 2
axes[1].hist(df['body_mass_g'], bins=30)
axes[1].set_xlabel('Body Mass')
axes[1].set_ylabel('Count')
axes[1].set_title('Distrubution of Penguin Body Mass')

# Space Adjustment
plt.subplots_adjust(left=0.1,
                    right=0.95,
                    bottom=0.1,
                    top=0.95,
                    wspace=0.5,
                    hspace=0.5)

plt.show()

```
